```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
 
%matplotlib inline
```

```python
flights = pd.read_csv('flights_data.csv')
print(flights.shape)
flights.head()
```

```python
flights.columns.unique()
```

```python
flights.info()
```

```python
base_color = sb.color_palette()[3]
sb.countplot(x='Airline',data=flights, color = base_color);
```

```python
base_color = sb.color_palette()[3]
sb.countplot(x='Airline',data=flights, color = base_color)
plt.xticks(rotation=90)
```

```python
base_color = sb.color_palette()[3]
sb.countplot(y='Airline',data=flights, color = base_color);
```

```python
 
sb.countplot(x='Source',data=flights)
plt.xticks(rotation=30)
plt.ylabel('Number of Flight', fontsize=12)
plt.xlabel('Source', fontsize=12)
```

```python
base_color = sb.color_palette()[2]
sb.countplot(x='Source',data=flights, color = base_color)
plt.xticks(rotation=30)
```

```python
base_color = sb.color_palette()[4]
gen_order = flights['Source'].value_counts().index
sb.countplot(x='Source',data=flights, color = base_color, order=gen_order)
```

```python
flights.isna().sum()
```

```python
na_counts = flights.isna().sum()
base_color = sb.color_palette()[5]
sb.barplot(na_counts.index.values, na_counts, color = base_color)
plt.xticks(rotation=90)
plt.ylabel('Number of missing values', fontsize=12)
```

```python
df = flights
sorted_counts = df['Destination'].value_counts()
plt.pie(sorted_counts, labels= sorted_counts.index, startangle=90, counterclock=False)
plt.axis('square')
plt.title('Flight Destination\'s')
```

```python
df = flights
sorted_counts = df['Destination'].value_counts()
plt.pie(sorted_counts, labels= sorted_counts.index, startangle=90, counterclock=False, wedgeprops={'width': 0.4})
plt.axis('square')
```

```python
plt.hist(data=df, x = 'Duration(minutes)')
```

```python
plt.hist(data=df, x = 'Price', bins = 20)
```

```python
bins= np.arange(0, df['Price'].max()+1, 1200)
plt.hist(data=df, x = 'Price', bins = bins)
plt.show()
```

```python
sb.distplot(df['Price']);
```

```python
 sb.distplot(df['Price'], kde=False);
```

```python
bin_edges = np.arange(0, df['Price'].max()+1, 1200)
sb.distplot((df['Price']), bins=bin_edges, kde=False, hist_kws = {'alpha' : 1});
```
